```python
# Задание 2
t = np.linspace(0, 10, num = 51, endpoint = True)
display(t)

f = np.cos(t)
display(f)

plt.plot(t,f, color = 'green')
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.show()
```

```python
# Задание 3
x = np.linspace(-3, 3, num = 51, endpoint = True)
display(x)

y1 = x ** 2
display(y1)

y2 = 2 * x + 0.5
display(y2)

y3 = -3 * x - 1.5
display(y3)

y4 = np.sin(x)
display(y4)

# Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием 
# fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке, 
# состоящей из двух строк и двух столбцов. В каждом графике массив x используется для координат по горизонтали.
# В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, 
# в правом нижнем - y4.Дайте название графикам: 'График y1', 'График y2' и т.д.
# Для графика в левом верхнем углу установите границы по оси x от -5 до 5.
# Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.
# Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.

fig, ax = plt.subplots(nrows = 2, ncols =2 )

ax[0,0].plot(x, y1)
ax[0,1].plot(x, y2)
ax[1,0].plot(x, y3)
ax[1,1].plot(x, y4)

ax[0,0].set_title('График y1')
ax[0,1].set_title('График y2')
ax[1,0].set_title('График y3')
ax[1,1].set_title('График y4')

ax[0,0].set_xlim(-5, 5)

plt.rcParams["figure.figsize"] = [8, 6]
plt.subplots_adjust(hspace = 0.3)
plt.subplots_adjust(wspace = 0.4) # специально ставим 0.4, при 0.3 заголовки нижней пары наползают на оси
# верхней пары
plt.show()
```

```python
# Задание 4:
import csv
import pandas as pd
matplotlib.style.use('fivethirtyeight')

df =  pd.read_csv('creditcard.csv.zip', delimiter = ",", compression='zip', header=0)
df
```

```python
# Посчитайте с помощью метода value_counts количество наблюдений для каждого значения целевой переменной 
# Class и примените к полученным данным метод plot, чтобы построить столбчатую диаграмму. Затем постройте 
# такую же диаграмму, используя логарифмический масштаб.
df['Class'].value_counts().plot(kind = 'bar')
plt.show()
df['Class'].value_counts().plot(kind = 'bar', logy = True)
plt.show()
```

```python
x1 = df.loc[df.Class==1, 'V1']
x2 = df.loc[df.Class==0, 'V1']

kwargs = dict(alpha=0.5, bins=20)

plt.hist(x1, **kwargs, color='red', label='Class 0', density = True)
plt.hist(x2, **kwargs, color='grey', label='Class 1', density = True)
plt.gca().set(title='Frequency Histogram of Diamond Depths', xlabel='Class')
#plt.xlim(50,75)
plt.legend();
```
